# Plotting

For more samples just go to the [github page](https://github.com/ColCarroll/strava_calendar)

```python
plotter = Plotter(myPlotterData)
```

## Daily Plotting

```python
year = 2019
month=10
day=20
sport="running"
label=None
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

myLineColour = 'cyan'
myBackgroundColour = 'gray'

plotter.plot_day( year = year, 
                 month = month, 
                 day = day, 
                 fig=fig, 
                 ax=ax,
                drawColor=myLineColour,
                backgroundColor=myBackgroundColour)
```

```python
plotter.plot_day( year=year, month = month, day = day, fig=fig, ax=ax,savePlot=True,fileName='outFile/Oct6.png')
```

## Monthly Plotting

```python
fig, ax = plt.subplots(figsize=(9, 6))

fig, ax, offset = plotter.plot_month(year=2019, 
                                     month=10, 
                                     fig=fig, 
                                     ax=ax,
                                     drawColor='cyan',
                                     backgroundColor='gray')
ax.text(0, offset + 4.2, 'Race Month!', fontdict={'fontsize': 32, 'fontweight': 'heavy'}, alpha=0.5)
```

## Yearly Plotting

```python
year = 2019
plot_size=1
n_cols=6
month_gap=0
col_gap=0.5
sport="running"
label=None
```

```python
fig, ax = plt.subplots(figsize=(plot_size * 5 * n_cols, plot_size * 40 / n_cols))
fig, ax = plotter.plot_year(year=year, 
                            fig=fig, 
                            ax=ax, 
                            n_cols=n_cols, 
                            month_gap=month_gap, 
                            col_gap=col_gap,
                            drawColor='cyan',
                            backgroundColor='gray')
if label is None:
    label = str(year)
ax.text(0, -1, label, fontdict={"fontsize": 32, "fontweight": "heavy"}, alpha=0.5)
```
